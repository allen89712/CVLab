```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# 
image1 = cv2.imread('C:/Users/User/Desktop/school/Cvml/DataSet/baboon.jpg')
image2 = cv2.imread('C:/Users/User/Desktop/school/Cvml/DataSet/lena.png')
if image1 is None:
    print("1")
else:
  # 
  blue_channel1, green_channel1, red_channel1 = cv2.split(image1)
  blue_channel2, green_channel2, red_channel2 = cv2.split(image2)
# 0255
blue_channel1 = np.clip(blue_channel1, 0, 255)
green_channel1 = np.clip(green_channel1, 0, 255)
red_channel1 = np.clip(red_channel1, 0, 255)

blue_channel2 = np.clip(blue_channel2, 0, 255)
green_channel2 = np.clip(green_channel2, 0, 255)
red_channel2 = np.clip(red_channel2, 0, 255)

channel_mean1 = np.mean(blue_channel1)
channel_std1 = np.std(blue_channel1)
channel_mean2 = np.mean(blue_channel2)
channel_std2 = np.std(blue_channel2)

channel_mean3 = np.mean(green_channel1)
channel_std3 = np.std(green_channel1)
channel_mean4 = np.mean(green_channel2)
channel_std4 = np.std(green_channel2)

channel_mean5 = np.mean(red_channel1)
channel_std5 = np.std(red_channel1)
channel_mean6 = np.mean(red_channel2)
channel_std6 = np.std(red_channel2)

blue_channel1 = ((blue_channel1 - channel_mean1) * (channel_std2 / channel_std1)) + channel_mean2
green_channel1 = ((green_channel1 - channel_mean3) * (channel_std4 / channel_std3)) + channel_mean4
red_channel1 = ((red_channel1 - channel_mean5) * (channel_std6 / channel_std5)) + channel_mean6

# 
processed_image = cv2.merge([blue_channel1, green_channel1, red_channel1])

# 
processed_image = cv2.convertScaleAbs(processed_image)
cv2.imwrite('processed_image.jpg', processed_image)
image_rgb = cv2.cvtColor(processed_image, cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb)
plt.show()
```
